```python
cols =['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long',
       'sqft_living15', 'sqft_lot15', 'year', 'month']
```

```python
df = df[cols]
```

```python
X = df.drop('price',axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 , random_state = 101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([Dense(19, activation='relu'),
                    Dense(19, activation='relu'),
                    Dense(19, activation='relu'),
                    Dense(1)])

model.compile(optimizer='adam', loss= 'mse')
```

```python
result = model.fit(X_train,y_train,validation_data=(X_test,y_test), batch_size=128, epochs=200)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot(figsize=(12,8))
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
def percentage_error(dataframe,pred, true_value):
    
    '''
    dataframe: Pass in pandas dataframe
    pred = dataframe['Predictions'] #model.predict(x_test)
    true_value = dataframe['True Values'] #y_test
    
    '''
    
    difference = true_value - pred
    
    dataframe['Error(%)']= np.absolute(difference/true_value*100)
    
    print('Average Percentage Error is: {}'.format(dataframe['Error(%)'].mean()))
```

```python
from sklearn.metrics import explained_variance_score,mean_absolute_error, mean_squared_error, mean_squared_log_error
```

```python
predictions = model.predict(X_test)
df_pred = pd.DataFrame() 
df_pred['True Values'] = y_test
df_pred['Predictions'] = predictions
```

```python
percentage_error(df_pred,df_pred['Predictions'],df_pred['True Values'])
```

```python
mean_absolute_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)
```
